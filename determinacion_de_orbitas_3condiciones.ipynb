## **CÁLCULO DE LOS ELEMENTOS ORBITALES**

Lo que sigue a continuación es relacionar los tres vectores posición con los elementos orbitales que se desean determinar. El producto cruz entre dos vectores heliocéntricos, digamos $\vec{H} = \vec{r_1} × \vec{r_2}$, permite definir un vector que es ortogonal al plano orbital. La división de este vector por su magnitud $(\hat{H} = \vec{H} \ / \ |\vec{H}|)$ permite obtener un vector unitario que también es ortogonal al plano orbital. Sean entonces $H_1$, $H_2$ y $H_3$ las componentes de dicho vector unitario.


```python
# VECTOR UNITARIO H
v1 = np.array([x1, y1, z1])
v2 = np.array([x2, y2, z2])
v3 = np.array([x3, y3, z3])
H = np.cross(v1, v2)
H1 = H[0]
H2 = H[1]
H3 = H[2]
Hnorm = np.linalg.norm(H)
print('---> r1 x r2 = H = ({}, {}, {}) \n\n---> |H| = {}'.format(H1, H2, H3, Hnorm))

h_gorro = H / Hnorm
B1 = h_gorro[0]
B2 = h_gorro[1]
B3 = h_gorro[2]

print('\n---> H_gorro = ({}, {}, {}) \n\n---> B1 = {} \n\n---> B2 = {} \n\n---> B3 = {}'.format(B1, B2, B3, B1, B2, B3))
```



---
Recordar aplicar criterio del cuadrante a $\Omega$


---



Pero ya hemos visto cómo es la relación entre este vector unitario ortogonal al plano orbital con los elementos orbitales $\Omega$ e  $i$  través de la ecuación. Por lo tanto:



> $$\hat{h} = \sin{\Omega}\sin{i} \ \hat{i}-\cos{\Omega}\sin{i} \ \hat{j} + \cos{i} \ \hat{k}$$

entonces:


> $$ i = \arccos{B_3}$$


> $$ \Omega = \arctan{\frac{B_1}{-B_2}}$$







```python
i = math.degrees(math.acos(B3)) # RECORDAR QUE SI i > 90 LA DIRECCIÓN ES EN SENTIDO DE LAS AGUJAS DEL RELOJ
Omega = math.degrees(math.atan2(B1, -B2)) # APLICANDO CRITERIO DEL CUADRANTE

print('---> inclinación: i = {}º \n---> longitud del nodo ascendente: Omega = {}º'.format(i, Omega))
```

Definiendo el vector $\hat{n}$ que se ubica en el plano $xy$ y dirigido en la dirección del nodo ascendente y $\hat{m}$ ortogonal a $\hat{n}$ pero dirigido en el plano de la órbita en que se mueve el objeto de la siguiente manera:



> $$\hat{n} = \cos{\Omega} \hat{i} + \sin{\Omega}\hat{j}$$



```python
n1 = math.cos(math.radians(Omega))
n2 = math.sin(math.radians(Omega))
n3 = 0
n_gorro = np.array([n1,n2,n3])
print('---> n_gorro = ({}, {}, {})'.format(n1, n2, n3))
```



> $$\hat{m} = -\sin{\Omega}\cos{i} \hat{i} + \cos{\Omega}\cos{i} \hat{j} + \sin{i}\hat{k}$$



```python
m1 = - math.sin(math.radians(Omega)) * math.cos(math.radians(i))
m2 = math.cos(math.radians(Omega)) * math.cos(math.radians(i))
m3 = math.sin(math.radians(i))
m_gorro = np.array([m1, m2, m3])
print('---> m_gorro = ({}, {}, {})'.format(m1, m2, m3))
```

La relación entre estas cantidades es:



> $$\hat{r} = \hat{n}\cos{(\omega + \theta)+\hat{m}\sin{(\omega + \theta)}}$$



De la ecuación de la órbita se tiene que:


> $$[(\vec{r_1}-\vec{r_2})·\hat{n}]e\cos\omega+[(\vec{r_1}-\vec{r_2})·\hat{m}]e\sin{\omega} ={r_2}-{r_1}$$





> $$[(\vec{r_1}-\vec{r_3})·\hat{n}]e\cos\omega+[(\vec{r_1}-\vec{r_3})·\hat{m}]e\sin{\omega} ={r_3}-{r_1}$$


Se tiene entonces un sistema de ecuaciones lineales 2X2

```python
# PRIMERO, CALCULAMOS LAS COMPONENTES DE LA RESTA DE LOS VECTORES
v1_2 = v1- v2
v1_3 = v1 - v3
print('---> r1 - r2 (vector) = ({}, {}, {}) u.a \n\n---> r1 - r3 (vector) = ({}, {}, {}) u.a'.format(v1_2[0], v1_2[1], v1_2[2], v1_3[0], v1_3[1], v1_3[2]))
```

```python
# LUEGO, SE CALCULA LA DIFERENCIA DE LAS NORMAS
r2_1 = r2 - r1
r3_1 = r3 - r1

print('r2 - r1 (normas) = {} u.a \n\n r3 -r1 (normas) = {} u.a'.format(r2_1, r3_1))
```

```python
# Y LOS COEFICIENTES DE LOS PRODUCTOS PUNTO:
c1 = v1_2 @ n_gorro
c2 = v1_2 @ m_gorro
c3 = v1_3 @ n_gorro
c4 = v1_3 @ m_gorro

print('---> c1 = {} \n\n---> c2 = {} \n\n---> c3 = {} \n\n---> c4 = {}'.format(c1, c2, c3, c4))
```

Por lo tanto, el sistema queda expresado de la siguiente manera:



>$$c_1e\cos{\omega}+c_2e\sin{\omega} = r_2 - r_1$$


> $$c_3e\cos{\omega}+c_4e\sin{\omega} = r_3 - r_1$$

Tomando como variables auxiliares $H = e\cos{\omega}$ y $K = e\sin{\omega}$, se resuelve el sistema:


>$$c_1H+c_2K = r_2 - r_1$$


> $$c_3H+c_4K = r_3 - r_1$$

```python
#SOLUCIÓN DEL SISTEMA 2X2

A = np.array([[c1, c2], [c3, c4]])
B = np.array([r2_1, r3_1])
sol = np.linalg.solve(A, B)
H = sol[0]
K = sol[1]
print('---> H = {} \n\n---> K = {}'.format(H, K))
```

```python
#VALORES DE LA EXCENTRICIDAD Y ARGUMENTO DE LATITUD DEL PERIHELIO

e = math.sqrt(H**2 + K**2)
omega = math.degrees(math.atan2(K, H)) # APLICANDO CRITERIO DEL CUADRANTE
print(math.degrees(math.tan(K/H)))

print('--> excentricidad: e = {} \n\n---> argumento de latitud del periherlio: omega = {}º'.format(e, omega))
```

```python
# FUNCIÓN PARA LA FECHA JULIANA
def fecha_juliana(fecha):
    year = fecha.year
    month = fecha.month
    day = fecha.day
    hora = fecha.hour
    minuto = fecha.minute
    segundo = fecha.second

    a = int((14 - month) / 12)
    y = year + 4800 - a
    m = month + 12 * a - 3

    # Calcula la fecha Juliana
    jdn = day + int((153 * m + 2) / 5) + 365 * y + int(y / 4) - int(y / 100) + int(y / 400) - 32045

    # Agrega la fracción del día
    fraccion = (hora - 12) / 24 + minuto / 1440 + segundo / 86400
    jd = jdn + fraccion

    return jd


""" t_1 """
tiene_fecha_juliana_t_1 = input("\n¿Tiene la fecha juliana de t_1? (s/n): ")

if tiene_fecha_juliana_t_1 == "s":
    # Si tiene la fecha juliana, t_1
    print('\n* t_1 = {}'.format(t_1))

else:
    # Si no tiene la fecha juliana, pedirle los datos necesarios
    año_1 = int(input("\naño de t_1 = "))
    mes_1 = int(input("\nmes de t_1 = "))
    dia_1 = int(input("\ndía de t_1 = "))
    hora_1 = int(input("\nhora de t_1 = "))
    minuto_1 = int(input("\nminuto de t_1 = "))
    segundo_1 = int(input("\nsegundo de t_1 = "))
    fecha_1 = datetime.datetime(año_1, mes_1, dia_1, hora_1, minuto_1, segundo_1)
    t_1 = fecha_juliana(fecha_1)
    print('\n* t_1 = {}'.format(t_1))

""" t_2 """
tiene_fecha_juliana_t_2 = input("\n¿Tiene la fecha juliana de t_2? (s/n): ")

if tiene_fecha_juliana_t_2 == "s":
    # Si tiene la fecha juliana, t_2
    print('\n* t_2= {}'.format(t_2))

else:
    # Si no tiene la fecha juliana, pedirle los datos necesarios
    año_2 = int(input("\naño de t_2 = "))
    mes_2 = int(input("\nmes de t_2 = "))
    dia_2 = int(input("\ndía de t_2 = "))
    hora_2 = int(input("\nhora de t_2 = "))
    minuto_2 = int(input("\nminuto de t_2 = "))
    segundo_2 = int(input("\nsegundo de t_2 = "))
    fecha_2 = datetime.datetime(año_2, mes_2, dia_2, hora_2, minuto_2, segundo_2)
    t_2 = fecha_juliana(fecha_2)
    print('\n* t_2 = {}'.format(t_2))

""" t_3 """
tiene_fecha_juliana_t_3 = input("\n¿Tiene la fecha juliana de t_3? (s/n): ")

if tiene_fecha_juliana_t_3 == "s":
    # Si tiene la fecha juliana, t_3
    print('\n* t_3= {}'.format(t_3))

else:
    # Si no tiene la fecha juliana, pedirle los datos necesarios
    año_3 = int(input("\naño de t_3 = "))
    mes_3 = int(input("\nmes de t_3 = "))
    dia_3 = int(input("\ndía de t_3 = "))
    hora_3 = int(input("\nhora de t_3 = "))
    minuto_3 = int(input("\nminuto de t_3 = "))
    segundo_3 = int(input("\nsegundo de t_3 = "))
    fecha_3 = datetime.datetime(año_3, mes_3, dia_3, hora_3, minuto_3, segundo_3)
    t_3 = fecha_juliana(fecha_3)
    print('\n* t_3 = {}'.format(t_3))
```

# ***MOVIMIENTO MEDIO (n)***

```python
# FUNCIONES PARA EL MOV. MEDIO

def n_sol(longitud_eje_mayor,proporcion):

    """ Calcula el movimiento medio EN GRADOS de un objeto celeste ORBITANDO EL SOL
        ---> La proporción es m2 / m_sol"""
    n = (180/math.pi) * 0.01720209908 * (math.sqrt(1+proporcion)) / longitud_eje_mayor**(3/2)
    return n

def n_masadespreciable_sol(longitud_eje_mayor):

    """ Calcula el movimiento medio EN GRADOS de un objeto celeste de MASA DESPRECIABLE ORBITANDO EL SOL"""

    n = (180/math.pi) * 0.01720209908 / longitud_eje_mayor**(3/2)
    return n

def n_general(longitud_eje_mayor, m1, m2):

    """ Calcula el movimiento medio EN GRADOS de un objeto celeste m2 ORBITANDO m1"""
    n = (180/math.pi) * math.sqrt((6.67e-11 * (86400**2) / (149597870.7e3)**3)* (m1+ m2)) / longitud_eje_mayor**(3/2)
    return n



mov_medio_sol = input("\n¿El cuerpo celeste se encuentra orbitando al Sol? (s/n): ")

if mov_medio_sol == "s":
    masa_despreciable = input("\n¿Puede considerarse de masa despreciable? (s/n): ")
    if masa_despreciable == "s":
        n = n_masadespreciable_sol(a)
        print("\n---> Movimiento medio: n = {} º/d".format(n))

    else:
        datos = int(input("\n¿Qué datos tiene?:\n\n  m2 / m_Sol (1) \n\n  m_Sol / m2 (2)\n "))
        if datos == 1:
            proporcion = float(input("\nIngrese m2 / m_Sol: "))
            n = n_sol(a, proporcion)
            print("\n---> Movimiento medio: n = {} º/d".format(n))
        else:
            p = float(input("\nIngrese m_Sol / m2: "))
            proporcion = 1/p
            n = round(n_sol(a, proporcion), 7)
            print("\n---> Movimiento medio: n = {} º/d".format(n))

else:
    m1 = float(input("\ningrese el valor de m1: "))
    m2 = float(input("\ningrese el valor de m2: "))
    n = n_general(a, m1, m2)
    print("\n---> Movimiento medio: n = {} º/d".format(n))

```
